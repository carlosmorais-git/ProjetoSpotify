### Usando rotas de paginas

https://reactrouter.com/home

- BrowserRouter: 
    -   Toda estrutura de rotas precisa esta envolvida por element, é um gerenciador de rotas

- Routes:
    -   Conjunto de rotas

- Route => Uma rota expecifica:
    - path = Caminho da url
    - element = Qual componente esta vinculado a esse caminho 
    - /:id = Esse parametro no final da URL indica que havera um valor dinamico( tenta pegar o texto passado ou qualquer outro apos esse :)

```python
<BrowserRouter>
    <Header />
    <Routes>
      <Route path="/" element={<Home />} />
      <Route path="/artists" element={<Artists />} />
      <Route path="/artist/:id" element={<Artist />} />
    </Routes>
</BrowserRouter>
```

Em Cluster faço a conexao



Para substituir o <db_password> vá em:

- Gerencie usuários do banco de dados:

    Vá até a aba Database Access (Acesso ao Banco de Dados).
    Você verá a lista de usuários cadastrados.


- Recupere ou redefina a senha:

    Se lembrar da senha, use-a no lugar de <db_password>.
    Se não lembrar, clique no usuário e selecione Edit para definir uma nova senha.
    

- Atualize a string de conexão:

    Depois de definir a senha, substitua <db_password> por ela na sua string de conexão.
